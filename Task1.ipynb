# **Comprehensive Sales Transactions Database**

```python
from google.colab import files
uploaded = files.upload()
```

```python
#Load Data
import pandas as pd
import numpy as np

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print("✅ Data Loaded Successfully\n")
print("Shape of dataset:", df.shape)
print("\n🔹 First 5 Rows:")
display(df.head())
```

```python
#Data Info
print("📊 Data Info:")
df.info()

print("\n📊 Summary Statistics:")
display(df.describe(include='all'))
```

```python
#Column Names
print("📌 Column Names:")
for col in df.columns:
    print("-", col)
```

```python
#Data Dictionary
data_dict = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes.astype(str)
})

print("📘 Data Dictionary:")
display(data_dict)

data_dict.to_csv("data_dictionary.csv", index=False)
```

```python
#Missing Values
print("🔍 Missing Values in Each Column:")
missing = df.isnull().sum()
display(missing)
```

```python
#Duplicate Check
duplicates = df.duplicated().sum()
print("🔁 Number of Duplicate Rows:", duplicates)
```

```python
#Remove Duplicates
df = df.drop_duplicates()

print("✅ After Removing Duplicates:")
print("New Shape:", df.shape)
```

```python
#Handle Missing Values
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

print("✅ Missing Values After Cleaning:")
display(df.isnull().sum())
```

```python
#Convert Date Columns
for col in df.columns:
    if "date" in col.lower() or "dob" in col.lower():
        df[col] = pd.to_datetime(df[col], errors='coerce')

print("📅 Data Types After Date Conversion:")
print(df.dtypes)
```

```python
#Feature Engineering
today = pd.Timestamp.today()

# Create Age if DOB exists
if 'dob' in df.columns:
    df['customer_age'] = (today - df['dob']).dt.days // 365
    print("✅ Customer Age Column Created")

# Extract Month from Purchase Date
for col in df.columns:
    if "purchase" in col.lower() and "date" in col.lower():
        df['purchase_month'] = df[col].dt.month
        print("✅ Purchase Month Column Created")

print("\n🔹 Updated Data:")
display(df.head())
```

```python
#Outlier Detection & Removal
numeric_cols = df.select_dtypes(include=np.number).columns

print("📊 Numeric Columns:", list(numeric_cols))

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    before = df.shape[0]

    df = df[(df[col] >= q1 - 1.5 * iqr) &
            (df[col] <= q3 + 1.5 * iqr)]

    after = df.shape[0]

    print(f"🔧 {col}: Removed {before - after} outliers")
```

```python
#Final Dataset Check
print("✅ Final Cleaned Dataset Shape:", df.shape)

print("\n🔹 Final Preview:")
display(df.head())

print("\n🔹 Final Info:")
df.info()
```

```python
#Save & Download Files
df.to_csv("cleaned_data.csv", index=False)

from google.colab import files
files.download("cleaned_data.csv")
files.download("data_dictionary.csv")

print("📁 Files downloaded successfully!")
```
